```python
import netCDF4 as nc
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "PAAAA2025021500+012.nc"
dataset = nc.Dataset(file_path)
list(dataset.variables)
```

```python
T = dataset.variables['ta'][:]
```

```python
level = 16
plt.figure(figsize=(12, 8))
plt.imshow(T[level, :, :], cmap='inferno', aspect='auto', origin='lower')
plt.colorbar(label='Temperature')
plt.title('Heatmap of Temperature at Level ' + str(level))
plt.show()
```
